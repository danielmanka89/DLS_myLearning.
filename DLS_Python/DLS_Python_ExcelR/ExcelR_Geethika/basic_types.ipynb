### Boolean Operations


```python
a = True
b = False

print(a and b)
print(a or b)
print(not a)
print(not b)

# Explanation:
# Here a and b evaluates to False because both operands are not True.
# a or b evaluates to True because at least one operand is True.
# not a evaluates to False because a is True.
# not b evaluates to True because b is False.
# The logical operators used are 'and', 'or', and 'not'.
# The data type of the variables a and b is bool (boolean).
```

```python
# Give me boolean operations examples like above for practice
x = 10
y = 20
print(x < y)      # True
print(x > y)      # False
print(x == y)     # False
print(x != y)     # True
print((x < y) and (x != y))  # True
print((x > y) or (x == y))   # False
print(not (x < y))  # False
print(not (x > y))  # True

# Explanation:
# The comparisons x < y, x > y, x == y, and x != y evaluate to True or False based on the values of x and y.
# The logical operations combine these comparisons using 'and', 'or', and 'not' to produce final boolean results.
# The data type of the variables x and y is int (integer).
# The results of the comparisons and logical operations are of type bool (boolean).
# This provides practice with boolean operations and comparisons in Python.
# The data types involved are int and bool.
# The logical operators used are 'and', 'or', and 'not'.
```

```python
# Mutable: Once created, can be changed.
# Immutable: Once created, cannot be changed.
# List is mutable, meaning it can be changed after its creation.
# Tuple is immutable, meaning it cannot be changed after its creation.
```

```python
# In list directory, we have methods like append, extend, insert, pop, clear, sort, reverse, count, index, remove, copy etc.

# l = [1,2,3]

# in List:
# append is allowed             # l.append(4)
# clear is allowed            # l.clear()
# extend is allowed           # l.extend([5,6])
# insert is allowed           # l.insert(1,10)
# pop is allowed              # l.pop()
# remove is allowed           # l.remove(2)
# sort is allowed             # l.sort()
# reverse is allowed          # l.reverse()
# count is allowed            # l.count(2)
# index is allowed            # l.index(3)
# copy is allowed             # l.copy()    
```

```python
# In Tuple, memory cannot be changed.

# t = (1,2,3)

# in Tuple:
# append is not allowed         # t.append(4)
# clear is not allowed          # t.clear()
# copy is not allowed           # t.copy()
# extend is not allowed         # t.extend([4,5])
# insert is not allowed         # t.insert(1,4)
# pop is not allowed            # t.pop()
# remove is not allowed         # t.remove(2)
# reverse is not allowed        # t.reverse()
# sort is not allowed           # t.sort()

# index is allowed              # t.index(2)
# count is allowed              # t.count(2)
```

### String

```python
# String
# A string is a sequence of characters enclosed in single, double, or triple quotes.

# String methods are similar to tuple methods since strings are also immutable.
```

```python
dir(str)
```

```python
s1 = 'hello world'

print(s1.capitalize())
# Here first letter of the string is converted to uppercase and the rest to lowercase.

print(s1.title())
# Here the first letter of each word in the string is converted to uppercase.

print(s1.upper())
# Here all letters in the string are converted to uppercase.
```

```python
l1 = [1,8,6]
l1.sort()
l1

# Here the list is sorted in ascending order.
```

```python
# s1 = "Hello World"
# s1.upper()
# Here s1 value is not changed because strings are immutable, but a new string is returned with all uppercase letters.
```

```python
# .upper(): This will convert all letters in the string to uppercase.

upp_s1 = s1.upper()
print(upp_s1)
print(s1)

# Here upp_s1 contains the uppercase version of s1, while s1 remains unchanged.
```

```python
# .lower(): This will convert all letters in the string to lowercase.

s2 = "I LOVE PYTHON"
s2.lower()

# Here all letters in the string are converted to lowercase.
```

```python
# .swapcase(): This will swap the case of each letter in the string.

s2.swapcase()
s2

# Here all uppercase letters are converted to lowercase and vice versa.
```

```python
s3 = "    Python"

# Here leading spaces are removed from the string.
```

```python
# lstrip(), rstrip(), strip(): These methods are used to remove whitespace from the beginning, end, or both ends of a string, respectively.

s3.lstrip(), s3.rstrip(), s3.strip()

# here:
# lstrip() removes leading whitespace
# rstrip() removes trailing whitespace
# strip() removes both leading and trailing whitespace
```

```python
# .split(): This will split a string into a list based on a specified delimiter.

s4 = "Hello, World! Welcome to Python programming."
s4.split()

# Here the string is split into a list of words based on whitespace.
# Each word becomes an element in the resulting list.
# The original string remains unchanged.
```

```python
# .split(', '): This will split a string into a list based on a specified delimiter.

s5 = 'Apple, Banana, Cherry, Grapes, Melon'
s5.split(', ')

# Here the string is split into a list of fruits based on the comma and space delimiter.
# Each fruit becomes an element in the resulting list.
# The original string remains unchanged.
```

```python
# .split(): This will split a string into a list based on a specified delimiter.

s5 = 'Apple|Grapes,Lemon|Melon|Mango'
s5.split('|')

# Here the string is split into a list of fruits based on the '|' delimiter.
# Each fruit becomes an element in the resulting list.
# The original string remains unchanged.
```

```python
fruits_lst = ['Apple', 'Grapes', 'Lemon', 'Melon', 'Mango']
fruits_lst

# Here we have a list of fruits stored in the variable fruits_lst.
# Each fruit is an element in the list.
```

```python
# ''.join(): This will join the elements of a list into a single string, with a specified separator.

' '.join(fruits_lst)

# Here the elements of the list fruits_lst are joined into a single string, with each element separated by a space.
```

```python
# ','.join(): This will join the elements of a list into a single string, with a specified separator.

', '.join(fruits_lst)

# Here the elements of the list fruits_lst are joined into a single string, with each element separated by a comma.
```

```python
# .replace(): This will replace a substring within a string with another substring.

s6 = 'Anarkali'
s6.replace('Anar', 'Champa')

# Here 'Anar' in the string is replaced with 'Champa', resulting in 'Champakali'.
```
